```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
soil_surf = pd.read_excel('../data/nsw-landscape-resilience.xlsx', sheet_name='soil surface condition')
print(soil_surf.shape)
soil_surf.head(3)
```

```python

pivot = pd.pivot_table(soil_surf,
                       values='TRANSACTION_UNITS',
                       index='Community',
                       columns=['LitterCov', 'Nutrients'],
                       aggfunc='sum',
                       margins=True)

#comment out next two lines to work with totals
# for column in pivot:
#     pivot[column] = pivot[column]/pivot['All']
pivot
```
